```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def buah(image_path):
    # Load image
    image = cv2.imread(image_path)

    # Convert image to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Convert image to HSV
    image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Define lower and upper thresholds for green color
    lower_green = np.array([36, 25, 25])
    upper_green = np.array([86, 255, 255])

    # Define lower and upper thresholds for red color
    lower_red1 = np.array([0, 70, 50])
    upper_red1 = np.array([10, 255, 255])
    lower_red2 = np.array([170, 70, 50])
    upper_red2 = np.array([180, 255, 255])

    # Define lower and upper thresholds for yellow color
    lower_yellow = np.array([20, 100, 100])
    upper_yellow = np.array([30, 255, 255])

    # Create masks for green, red, and yellow colors
    mask_green = cv2.inRange(image_hsv, lower_green, upper_green)
    mask_red1 = cv2.inRange(image_hsv, lower_red1, upper_red1)
    mask_red2 = cv2.inRange(image_hsv, lower_red2, upper_red2)
    mask_red = cv2.bitwise_or(mask_red1, mask_red2)
    mask_yellow = cv2.inRange(image_hsv, lower_yellow, upper_yellow)

    # Apply masks to the original image
    result_green = cv2.bitwise_and(image_rgb, image_rgb, mask=mask_green)
    result_red = cv2.bitwise_and(image_rgb, image_rgb, mask=mask_red)
    result_yellow = cv2.bitwise_and(image_rgb, image_rgb, mask=mask_yellow)

    # Create a figure and axes
    fig, axes = plt.subplots(1, 4, figsize=(16, 4))

    # Display the original image
    axes[0].imshow(image_rgb)
    axes[0].set_title('asli')

    # Display the green color detection result
    axes[1].imshow(result_green)
    axes[1].set_title('Hijau')

    # Display the red color detection result
    axes[2].imshow(result_red)
    axes[2].set_title('Merah')

    # Display the yellow color detection result
    axes[3].imshow(result_yellow)
    axes[3].set_title('Kunig')

    # Show the plot
    plt.tight_layout()
    plt.show()
```

```python
gambar = "alfisyam.jpg"
```

```python
buah(gambar)
```
